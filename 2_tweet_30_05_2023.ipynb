Average True Range is a useful volatility indicator that tracks the average of true ranges over a typical 14-day period. 

On the #OpenBB SDK, ATR data can make for a rather useful covariate to your forecast model. Check it out!

```python
data = openbb.stocks.load("BP")
data = openbb.forecast.atr(data, close_column = 'Adj Close', high_column = 'High', low_column = 'Low')

openbb.forecast.regr_chart(data, target_column = "Close", past_covariates = 'true_range')
```
