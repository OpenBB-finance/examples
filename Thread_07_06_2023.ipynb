In the #OpenBB SDk, under the Stocks module is an extra Quantitative Analysis sub-module.

What does it do?



1/6


Beta, a measure of volatility of a stock with respect to the overall market, is one of the metrics available here.

Check out the scatter plot and accompanying regression plot.

2/6


The Capital Assets Pricing Model (CAPM) represents how markets value securities and hence, their expected return.

The values in the output represent the Beta value and systematic risk of the stock.

3/6


Risk calculations using Fama-French data from Dartmouth's researchers Eugene Fama and Kenneth R. French are also possible within this sub-module.

The #OpenBB SDK provides up-to-date raw Fama French data on all major factors as laid out by the researchers.


4/6


Finally, filtered data on the monthly performance of a ticker going back 5 years is available through this easy line of code.

The data can be plotted for cleaner results. Check it out!

And there you have it, a final cherry on the cake that is the Stocks module of the #OpenBB SDK.
