```python
# Load your OHLC data
data = obb.equity.price.historical("AAPL").to_df()

# Convert index (date) to a numerical value (e.g., days since start)
data.reset_index(inplace=True)
X = data.index.values.reshape(-1, 1)  # Now X is just numerical indices
y = data['close'].values

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Predict and plot
data['Predictions'] = model.predict(X)
plt.plot(data['close'], label='Actual Prices')
plt.plot(data['Predictions'], label='Predicted Prices')
plt.legend()
plt.show()
```
